行うことは、入力データに対して<br>
•daysInYearとdaysInMonthとdaysInWeekについて、日付から曜日を求めるプログラム<br>


①まず、daysInYearとdaysInMonthから、そのようなカレンダーが不正かどうかを判定する。<br>
②次に、与えられた日付がそのカレンダーに含まれうるかを判定する<br>
③最後に、1年目ではそれまでに蓄積されてきた余った日はなく、0001-01-01はA曜日だとしたとき、与えられた日付が何曜日かを計算する<br>

注意なのが閏月で、年日数を月日数で割った余りの合計が入力データだけ回る間に月日数上回るたびにその年は月を+1して考える<br>
次の閏月はあまり合計の上回った分からまた合計を数えて、同様に行う。

つまり、与えられた日付データを一回、daysInYear daysInMonth daysInWeek の条件のもとで、「年、月、日」で分解する<br>
入力データはこんな感じ(100 10 10 0001-01-10) →(100:日/年)　 (10:日/月)  (10:日/週)　(与えられる日付:0001-01-10)<br>


## <大まかにやること>
•**まずは入力される日付をsplitで日/年、日/月、日/週を変数に格納する**<br>
•**閏年判定してから、max月(13or12)、max日を変数化して、入力データの月と日が存在しうるか判定**<br>
•**(日/月)x(月/年)+残り日数=入力データに対するいままでの日数 → これを日/週で割って余りで曜日判定**<br>
