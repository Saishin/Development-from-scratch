```python
import numpy as np
import matplotlib.pyplot as plt

#relu関数
def ReLU(x):
    y = np.maximum(0, x)
    return y

def Sigmoid(x):
    return 1.0 / (1.0 + np.exp(-x))

#-8から8までの値0.1刻みで生成
X = np.arange(-5, 5, 0.1)
#Xの値をReLU関数へ渡す
Y1 = ReLU(X)
Y2 = Sigmoid(X)
#ReLuプロット
plt.subplot(1,2,1)
plt.plot(X,Y1)
plt.title("ReLu")
plt.xlim(-1,1)
plt.ylim(0, 1.0)

##Sigmoidプロット
plt.subplot(1,2,2)
plt.plot(X,Y2)
plt.title("Sigmoid")
plt.xlim(X.min(), X.max())
plt.ylim(0, 1.0)


#グラフ出力
plt.show()
```
